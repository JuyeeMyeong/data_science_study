```python
import pandas as pd
import numpy as np
```

```python
reg = pd.read_csv('../data/regression.csv')
```

```python
reg.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = reg["X"]
y = reg["Y"]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(X, y, 'o')
plt.show()
```

```python
line_fitter = LinearRegression() 
line_fitter.fit(X.values.reshape(-1,1), y)
```

```python
line_fitter.predict([[70]]) #X값을 넣어서 y값 예측
line_fitter.coef_ #기울기
```

```python
line_fitter.intercept_ #절편
```

```python
plt.plot(X, y, 'o')
plt.plot(X,line_fitter.predict(X.values.reshape(-1,1)))
plt.show() #기존 X 값으로 y를 예측
```
